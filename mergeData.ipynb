# *Health Data*
## General Information
- *SEQN*: Unique participant identifier for the dataset.
- *SDDSRVYR*: Survey participant ID year.

## Demographics
- *RIDSTATR*: Interview/Examination status.
- *RIAGENDR*: Gender (Male/Female).
- *RIDAGEYR*: Age in years at the time of screening.
- *RIDAGEMN*: Age in months at the time of screening.
- *RIDRETH1*: Race/ethnicity category.
- *RIDRETH3*: Another race/ethnicity classification.

## Examination
- *RIDEXMON*: Month of examination.
- *RIDEXAGM*: Age in months at the time of examination.
- *DMQMILIZ*: Military service status (Yes/No).

## Citizenship and Country of Origin
- *DMDBORN4*: Country of birth.
- *DMDCITZN*: Citizenship status.
- *DMDYRSUS*: Number of years in the U.S.

## Education and Marital Status
- *DMDEDUC3*: Education level, detailed.
- *DMDEDUC2*: Education level, broad.
- *DMDMARTL*: Marital status.

## Pregnancy
- *RIDEXPRG*: Pregnancy status at the time of examination.

## Language and Proxy Use
- *SIALANG*: Language of the survey.
- *SIAPROXY*: Whether a proxy answered the survey.
- *SIAINTRP*: Whether an interpreter was used during the survey.

## Family Language and Proxy Use
- *FIALANG*: Family interview language.
- *FIAPROXY*: Whether a proxy was used for the family interview.
- *FIAINTRP*: Whether an interpreter was used for the family interview.

## Medical Language and Proxy Use
- *MIALANG*: Medical interview language.
- *MIAPROXY*: Whether a proxy was used for the medical interview.
- *MIAINTRP*: Whether an interpreter was used for the medical interview.

### Adult Language Use
- *AIALANGA*: Adult language use status.

### Household Composition
- *DMDHHSIZ*: Total number of people in the household.
- *DMDFMSIZ*: Family size.
- *DMDHHSZA, **DMDHHSZB, **DMDHHSZE*: Other household composition details.

## Household Demographics
- *DMDHRGND*: Gender of household head.
- *DMDHRAGZ*: Age of household head.
- *DMDHREDZ*: Education level of household head.
- *DMDHRMAZ*: Marital status of household head.
- *DMDHSEDZ*: Household head employment status.

## Weights and Sampling
- *WTINT2YR* : Interview sample weight.
- *WTMEC2YR*: MEC sample weight.
- *SDMVPSU*: Variance estimation PSU.
- *SDMVSTRA*: Variance estimation strata.

## Income
- *INDHHIN2*: Household income.
- *INDFMIN2*: Family income.
- *INDFMPIR*: Income to poverty ratio.

## Dietary Information
- *WTDRD1_x, **WTDR2D_x*: Dietary day one and two weights.
- *DR1ILINE, **DR2ILINE*: Line number in dietary recall.
- *DR1DRSTZ, **DR2DRSTZ*: Dietary recall status.
- *DR1EXMER, **DR2EXMER*: Whether breakfast was reported.
- *DRABF_x, **DRABF_y*: Alcohol consumption.
- *DRDINT_x, **DRDINT_y*: Dietary interview status.

## Dietary Recall

- *DR1DAY, **DR2DAY*: Day of the week.
- *DR1LANG, **DR2LANG*: Language of dietary interview.

```python
import pandas as pd

# Load dataframes
demographic_df = pd.read_sas('demographic data.XPT', format='xport')
dietary1_df = pd.read_sas('Dietary data 1.XPT', format='xport')
dietary2_df = pd.read_sas('Dieatary data 2.XPT', format='xport')
GHB_Jdf = pd.read_sas('GHB_J.XPT', format='xport')
HSCRP_Jdf = pd.read_sas('HSCRP_J.XPT', format='xport')
```

```python
dietary_df = dietary1_df.merge(dietary2_df, on='SEQN', how='outer')
dietary_df = dietary_df.drop_duplicates(subset='SEQN')
dietary_df
```

```python
HSCRP_Jdf_GHB_Jdf = HSCRP_Jdf.merge(GHB_Jdf, on='SEQN', how='outer')
HSCRP_Jdf_GHB_Jdf = HSCRP_Jdf_GHB_Jdf.drop_duplicates(subset='SEQN')
HSCRP_Jdf_GHB_Jdf
```

```python
merged_df = dietary_df.merge(HSCRP_Jdf_GHB_Jdf, on='SEQN', how='outer')
merged_df = merged_df.drop_duplicates(subset='SEQN')
merged_df
```

```python
merged_df = demographic_df.merge(merged_df, on='SEQN', how='outer')
merged_df = merged_df.drop_duplicates(subset='SEQN')
merged_df
```

```python
# Create a list of all column names
columns_list = list(merged_df.columns)
print(columns_list)
```

```python
# Save merged dataframe to a new file
merged_df.to_csv('merged_health_data.csv', index=False)
```
